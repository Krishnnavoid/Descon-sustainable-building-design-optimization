5. Write a function definition named is_odd that takes in a number and returns True or False if that number is odd.

```python
def is_odd(number):
    return number % 2 != 0

print(is_odd(5))
print(is_odd(10))
```

6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(numbers):
    return max(set(numbers), key=numbers.count)

print(mode([1, 2, 2, 3, 4, 2, 5]))
```

Exercise 3: Sort following NumPy array \\
Case 1: Sort array by the second row \\
Case 2: Sort the array by the second column

```python
arr = np.array([[5, 2, 8],
                [3, 7, 6],
                [1, 4, 9]])

# Case 1: Sort by second row
sorted_by_row = arr[:, arr[1].argsort()]
print("Sorted by second row:\n", sorted_by_row)

# Case 2: Sort by second column
sorted_by_column = arr[arr[:, 1].argsort()]
print("Sorted by second column:\n", sorted_by_column)
```

Exercise 1: Read Total profit of all months and show it using a line plot \\
Total profit data provided for each month. Generated line plot must include the following properties: –

```python
import matplotlib.pyplot as plt
```

```python
!wget https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv
sales_df = pd.read_csv('/content/company_sales_data.csv')

plt.plot(sales_df['month_number'], sales_df['total_profit'])
plt.xlabel("Month Number")
plt.ylabel("Total profit")
plt.title("Total Profit Per Month")
plt.show()
```

Exercise : Read face cream and facewash product sales data and show it using the bar chart \\
The bar chart should display the number of units sold per month for each product. \\
Add a separate bar for each product in the same chart.

```python
plt.bar(sales_df['month_number'] - 0.2, sales_df['facecream'], width=0.4, label='Face Cream')
plt.bar(sales_df['month_number'] + 0.2, sales_df['facewash'], width=0.4, label='Face Wash')

plt.xlabel("Month Number")
plt.ylabel("Units Sold")
plt.title("Face Cream and Face Wash Sales")
plt.legend()
plt.show()
```
