2. Print multiplication table from 1 to 10

```python
import numpy as np

nums = np.linspace(1, 10, 10,dtype=int)

for i in nums:
    print(f"Multiplication table of {i}")
    for j in nums:
        print(f"{i} x {j} = {i*j}")
    print()
```

6. Write a function definition named mode that takes in sequence of numbers and returns the most commonly occuring value

```python
def mode(seq):
    max_count = 0
    mode_value = None

    for num in seq:
        count = seq.count(num)
        if count > max_count:
            max_count = count
            mode_value = num

    return mode_value
```
